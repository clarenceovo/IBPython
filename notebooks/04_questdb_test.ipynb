# QuestDB Test
Creates the OHLCV table and queries latest bars.

```python
import pandas as pd
from src.config.settings import load_settings
from src.transport.questdb_client import QuestDBClient

settings = load_settings(env_file='../.env')
async with QuestDBClient(host=settings.questdb_host, port=settings.questdb_port, user=settings.questdb_user, password=settings.questdb_password, database=settings.questdb_database) as qdb:
    await qdb.create_market_ohlcv_table()
    rows = await qdb.query_latest_bars(limit=20)
display(pd.DataFrame(rows))
```
