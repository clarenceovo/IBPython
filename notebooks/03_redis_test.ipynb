# Redis Test
Writes and reads one latest-bar payload.

```python
from datetime import datetime, timezone
from src.config.settings import load_settings
from src.feeds.models import OHLCVBar
from src.transport.redis_client import MarketDataRedisClient

settings = load_settings(env_file='../.env')
bar = OHLCVBar(symbol='SPY', asset_class='equity', exchange='SMART', currency='USD', timestamp=datetime.now(timezone.utc), open=1, high=2, low=1, close=1.5, volume=100, bar_size='1 min')
async with MarketDataRedisClient(settings.redis_url) as redis:
    key = await redis.set_latest_bar(bar)
    loaded = await redis.get_latest_bar('equity', '1 min')
    print(key)
    print(loaded)
```
