# Index Composition Sync
Demonstrates provider abstraction with a static payload.

```python
from src.config.settings import load_settings
from src.feeds.index_composition import IndexCompositionService, StaticIndexCompositionProvider
from src.transport.redis_client import MarketDataRedisClient

settings = load_settings(env_file='../.env')
provider = StaticIndexCompositionProvider({'SPX': [{'symbol': 'AAPL', 'name': 'Apple Inc.', 'weight': 0.07, 'sector': 'Technology', 'exchange': 'NASDAQ', 'currency': 'USD'}]})
async with MarketDataRedisClient(settings.redis_url) as redis:
    service = IndexCompositionService(provider, redis)
    payload = await service.sync_index('SPX')
    print(payload.model_dump())
```
