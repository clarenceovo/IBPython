# Load OHLCV
Loads IBKR historical bars and displays a DataFrame.

```python
import pandas as pd
from mpmath import rand

from src.config.settings import load_settings
from src.feeds.ibkr_feed import IBKRFeedClient
from src.feeds.models import OHLCVRequest
from random import randint
settings = load_settings(env_file='../.env')
request = OHLCVRequest(symbol='SPY', asset_class='equity', exchange='SMART', currency='USD', duration='1 D', bar_size='1 min')
client = IBKRFeedClient(settings.ibkr_host, settings.ibkr_port, randint(0,99))
try:
    await client.connect()
    bars = await client.load_historical_ohlcv(request)
    display(pd.DataFrame([bar.model_dump() for bar in bars]).tail())
finally:
    await client.disconnect()
```
